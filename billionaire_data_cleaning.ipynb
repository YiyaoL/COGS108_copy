```python
import requests  
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# read in csv file
df_college = pd.read_csv('billionaire_scraped.csv')

# remove data of billionaires without a college degree
df_college = df_college.dropna(subset=['Education_info'])
df_college = df_college[df_college['Education_info'].str.contains('Bachelor|Drop Out')]
df_college
```

```python
# group by college and rank by the number of billionaire alumni in descending order
grouped_df = df_college.groupby('Educations').count().sort_values(by='Name', ascending=False)
grouped_df
```

```python
# delete nonimportant columns
grouped_df = grouped_df.assign(num_alumni=grouped_df['Name']).assign(drop_out=grouped_df['dropout']).drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'Name', 'NetWorth', 'Country', 'Source', 'Rank', 'Industry', 'urls', 'Education_info', 'dropout'])
grouped_df
```
